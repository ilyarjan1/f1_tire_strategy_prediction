## 0) Setup


```python

# !pip install -q -r requirements.txt
import json
from pathlib import Path

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```


## 1) Configuration


```python

DATA_PATH = "data/f1_strategy.csv"
TARGET_COL = "pit_stop"
ARTIFACTS_DIR = Path("artifacts")
FIG_DIR = Path("figures")
ARTIFACTS_DIR.mkdir(exist_ok=True)
FIG_DIR.mkdir(exist_ok=True)
```


## 2) Load Data + Quick Profile


```python

df = pd.read_csv(DATA_PATH)
assert TARGET_COL in df.columns, f"Missing target column: {TARGET_COL}"

print("Shape:", df.shape)
print("Target classes:")
print(df[TARGET_COL].value_counts(dropna=False))
df.head()
```


## 3) EDA Plots


```python

num_df = df.select_dtypes(include=[np.number])
if not num_df.empty:
    num_df.hist(figsize=(14, 10), bins=25)
    plt.suptitle("Numeric Distributions", y=1.02)
    plt.tight_layout()
    plt.savefig(FIG_DIR / "numeric_distributions.png", dpi=150)
    plt.show()

if not num_df.empty:
    plt.figure(figsize=(10, 8))
    sns.heatmap(num_df.corr(numeric_only=True), cmap="coolwarm", center=0)
    plt.title("Correlation Heatmap")
    plt.tight_layout()
    plt.savefig(FIG_DIR / "correlation_heatmap.png", dpi=150)
    plt.show()
```


## 5) Review Training Outputs


```python

history_path = ARTIFACTS_DIR / "run_history.csv"
best_path = ARTIFACTS_DIR / "best_model_meta.json"

if history_path.exists():
    history_df = pd.read_csv(history_path).sort_values(["macro_f1", "accuracy"], ascending=False)
    display(history_df)
else:
    print("No run_history.csv yet. Run training first.")

if best_path.exists():
    best_meta = json.loads(best_path.read_text())
    print("Best model metadata:")
    print(json.dumps(best_meta, indent=2))
else:
    print("No best_model_meta.json yet. Run training first.")
```
